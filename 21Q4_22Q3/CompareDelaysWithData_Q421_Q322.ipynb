# Get Delays for a Train or Route

```python
# Get the stats we need from the files
def getDelays(folder, startCity, endCity, trainNumber = -1, excludeTrains = []):
  
    delaysByTrain = { }
    
    # Look through all trains
    for filename in os.listdir(folder):
    
        train = filename.find("Train")
        i = filename[train+5:len(filename)-4]
        i = int(i)
        
        filepath = folder + "/" + filename
        file = open(filepath, "r")
        text = file.read()

        delaysByTrain[i] = []
        
        toFind = "</span></td><td style=\"text-align: center;\">"
        length = len(toFind)
        searching = text.find(toFind)
        while (searching != -1) :
            end = text.find("<", searching + length)
            number = int(text[searching+length:end])
            if number > -130:
                delaysByTrain[i].append(number)
            searching = text.find(toFind, end)
    
    if trainNumber <= 0:
        for x in routeTrains:
            if x[0] == startCity and x[1] == endCity:
                listDelays = []
                for y in routeTrains[x]:
                    if y not in excludeTrains:
                        listDelays = listDelays + delaysByTrain[y]
    else: 
        listDelays = delaysByTrain[trainNumber]
        
    return listDelays
```

# Functions to Compare Delays

```python
def getlessDelayed(train1Delays, train2Delays, minutes):
    
    train1Delays.sort(reverse=True)
    train2Delays.sort(reverse=True)

    count = len(train1Delays) * len(train2Delays);
    countLess = 0;
    train2Pos = 0;

    for train1Pos in range(len(train1Delays)):
        countLess = countLess + train2Pos
        while(train2Pos < len(train2Delays) and (train1Delays[train1Pos] + minutes) < train2Delays[train2Pos]):
            countLess = countLess + 1
            train2Pos = train2Pos + 1
            
    return countLess / count * 100
```

```python
def getMoreDelayed(train1Delays, train2Delays, minutes):
    
    train1Delays.sort()
    train2Delays.sort()

    count = len(train1Delays) * len(train2Delays);
    countMore = 0;
    train2Pos = 0;

    for train1Pos in range(len(train1Delays)):
        countMore = countMore + train2Pos
        while(train2Pos < len(train2Delays) and train1Delays[train1Pos] > (train2Delays[train2Pos] + minutes)):
            countMore = countMore + 1
            train2Pos = train2Pos + 1

    return countMore / count * 100
```

```python
def getEqualDelayed(train1Delays, train2Delays):

    train1Delays.sort()
    train2Delays.sort()

    count = len(train1Delays) * len(train2Delays);
    countEqual = 0;
    train2PosStart = 0;
    train2Pos = 0;
    lastTrain2 = -1

    for train1Pos in range(len(train1Delays)):
        train2Pos = train2PosStart
        while(train2Pos < len(train2Delays) and train1Delays[train1Pos] >= train2Delays[train2Pos]):
            if (lastTrain2 != train2Delays[train2Pos]):
                lastTrain2 = train2Delays[train2Pos]
                train2PosStart = train2Pos;
            if (train1Delays[train1Pos] == train2Delays[train2Pos]):
                countEqual = countEqual + 1
            train2Pos = train2Pos + 1   
    
    return countEqual / count * 100
```
